```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

### Distribution Plot For Numeric Column

```python
sns.displot(df['Age'], kde = True)
```

```python
sns.distplot(df['Age'], kde = False, hist = True, rug = True )
```

```python
sns.distplot(df['Age'], bins = 10)
```

```python
sns.displot(df['RestingBP'], kde = 'False', color = 'blue')
```

```python
sns.distplot(df['RestingBP'], kde = 'False', color = 'blue')
```

```python
sns.distplot(df['Cholesterol'], kde = 'False', color = 'blue')
```

```python
sns.distplot(df['MaxHR'], kde = 'False', color = 'blue')
```

### Pie Chart on Categorical Data

```python
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.0f')
```

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.0f')
```

```python
df.groupby('RestingECG').size().plot(kind = 'pie', autopct = '%.0f')
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.0f')
```

### Violin Plot : Generally used when too much columns are used


```python
sns.violinplot(df['Age'])
```

```python
sns.violinplot(x =df['HeartDisease'], y = df['Sex'])
```

```python
sns.violinplot(x =df['HeartDisease'], y = df['Age'])
```

```python
sns.violinplot(x =df['HeartDisease'], y = df['Cholesterol'])
```
